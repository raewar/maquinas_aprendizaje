# Pregunta 2

Al tener las preguntas separadas, cada una con su archivo respectivo, se deben extraer los datos nuevamente.

```python
import pandas as pd
import numpy as np

url = "http://statweb.stanford.edu/~tibs/ElemStatLearn/datasets/prostate.data"
df = pd.read_csv(url, sep="\t", header=0)
df = df.drop("Unnamed: 0", axis=1)
istrain_str = df["train"]
istrain = np.asarray([True if s == "T" else False for s in istrain_str])
istest = np.logical_not(istrain)
df = df.drop("train", axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
df_scaled["lpsa"] = df["lpsa"]
```

```python
import sklearn.linear_model as lm

X = df_scaled.ix[:,:-1]
N = X.shape[0]
X.insert(X.shape[1], "intercept", np.ones(N))
y = df_scaled["lpsa"]
Xtrain = X[istrain]
ytrain = y[istrain]
Xtest = X[np.logical_not(istrain)]
ytest = y[np.logical_not(istrain)]
```

Se crean dos funciones: La primera, llamada MIS_order, no recibe parámetros de entrada y retorna una lista de tuplas. La función se encarga de calcular los Mutual Information Score para cada columna del DataFrame, y retornar finalmente una lista de tuplas las cuales consisten en (nombre_columna, Mi_Score_columna), esta lista se ordena en función a los Scores obtenidos, y finalmente es retornada.

La segunda funcion llamada plot_errors, recibe como parámetros una lista con errores de entrenamiento, una lista con errorres de prueba, y el título que quiera darse al gráfico. Esta función crea un gráfico que muestra tanto errores de prueba y de entrenamiento juntos, para asi poder compararlos.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn.metrics as met

def MIS_order():
    ordered_candidates={}
    order_index=[]
    for col in Xtrain:
        ordered_candidates[met.mutual_info_score(Xtrain[col],ytrain)]=col
        order_index.append((col,met.mutual_info_score(Xtrain[col],ytrain)))
    ss=sorted(order_index,key=lambda x: x[1])[::-1]
    return ss

def plot_errors(mse_train,mse_test,plot_title):
    plt.plot(range(1,len(mse_train)+1),mse_train,label="Error de entrenamiento",marker="o")
    plt.plot(range(1,len(mse_test)+1),mse_test,label="Error de pruebas",marker="*")
    plt.xlabel("Numero de variables en el modelo")
    plt.ylabel("Error cuadratico medio")
    plt.title(plot_title)
    plt.legend()
    if ("BSS" in plot_title):
        aux=range(1,9)[::-1]
        plt.xticks(range(1,9),aux)
    plt.show()
    return
```

Se desarrolla la función fss, la cual recibe como parámetros matrices de entrenamiento y prueba, asi como sus target respectivos. Ademas recibe una lista con las etiquetas de las columnas.

Esta función agrega en cada iteración una columna para ajustar al modelo de regresión lineal, se itera nuevamente entre todos los candidatos restantes para escoger al mejor entre ellos y agregarlo al modelo.

En cada iteración del algoritmo se calculan los Errores Cuadráticos Medios tanto de los datos de entrenamiento como los de prueba, y se almacena esta información en listas que al final del algoritmo son retornadas.

```python
def fss(x, y_train, x_test, y_test, names_x, k = 10000):
    p = x.shape[1]-1
    k = min(p, k)
    names_x = np.array(names_x)
    remaining = range(0, p)
    selected = [p]
    current_score = 0.0
    best_new_score = 0.0
    mse_train=[]
    mse_test=[]
    while remaining and len(selected)<=k :
        score_candidates = []
        for candidate in remaining:
            model = lm.LinearRegression(fit_intercept=False)
            indexes = selected + [candidate]
            x_train = x[:,indexes]
            mod=model.fit(x_train, y_train)
            
            predictions_train = mod.predict(x_train)
            predictions_test = mod.predict(x_test[:,indexes])
            
            residuals_train = predictions_train - y_train
            residuals_test = predictions_test - y_test
            
            mse_candidate = np.mean(np.power(residuals_train, 2))
            mse_test_candidate = np.mean(np.power(residuals_test, 2)) 
                  
            score_candidates.append((mse_candidate, mse_test_candidate, candidate))
       
        score_candidates.sort()
        score_candidates[:] = score_candidates[::-1]
        best_new_score,best_test_new_score ,best_candidate = score_candidates.pop()
        remaining.remove(best_candidate)
        selected.append(best_candidate)
        #print "selected = %s ..."%names_x[best_candidate]
        #print "totalvars=%d, mse = %f"%(len(indexes),best_new_score)
        
        mse_train.append(best_new_score)
        mse_test.append(best_test_new_score)

    return selected, mse_train, mse_test
```

El algoritmo BSS es muy similar al anterior, con la diferencia de que en este caso se inicia con un modelo ajustado por todas las columnas de la matriz de entrenamiento, y en cada iteración se va eliminando el peor candidato restante.

Al igual que el algoritmo FSS, esta función retorna listas con los errores de prueba y errores de entrenamiento, asi como el orden en que fueron eliminadas las características.

```python
def bss(x, y_train, x_test, y_test, names_x, k = 10000):
    p = x.shape[1]-1
    k = min(p, k)
    names_x = np.array(names_x)
    remaining = range(0, p)
    selected = range(0,p+1)
    deleted = [p]
    current_score = 0.0
    best_new_score = 0.0
    mse_train=[]
    mse_test=[]
    while remaining and len(selected)>1 :
        score_candidates = []
        for candidate in remaining:
            model = lm.LinearRegression(fit_intercept=False)
            indexes = selected[:]
            indexes.remove(candidate)
            x_train = x[:,indexes]
            mod=model.fit(x_train, y_train)
            
            predictions_train = mod.predict(x_train)
            predictions_test = mod.predict(x_test[:,indexes])
            
            residuals_train = predictions_train - y_train
            residuals_test = predictions_test - y_test
            
            mse_candidate = np.mean(np.power(residuals_train, 2))
            mse_test_candidate = np.mean(np.power(residuals_test, 2)) 
                  
            score_candidates.append((mse_candidate, mse_test_candidate, candidate))
       
        score_candidates.sort()
        score_candidates[:] = score_candidates[::-1]
        
        worst_new_score,worst_test_new_score ,worst_candidate = score_candidates.pop()
        remaining.remove(worst_candidate)
        selected.remove(worst_candidate)
        deleted.append(worst_candidate)
        #print "selected = %s ..."%names_x[worst_candidate]
        #print "totalvars=%d, mse = %f"%(len(indexes),worst_new_score)
        
        mse_train.append(worst_new_score)
        mse_test.append(worst_test_new_score)

    return deleted, mse_train, mse_test
```

Se debe escoger un nuevo criterio para saber cual es el próximo candidato a agregar al ajuste del modelo, para esto se crean las funciones fss_con_MSI y bss_con_MSI.

Ambos algoritmos recibne como parámetros las matrices de test y training, asi como sus respectivos targets. Ademas, reciben como parámetro una lista con las columnas ordenadas por su Mutual Information Score, la cual es extraida del algoritmo MIS_order mencionado anteriormente.

El funcionamiento de ambos algoritmos es similar a sus pares mencionados anteriormente, con la diferencia que esta vez, mediante el nuevo criterio, ya se sabe de antemano cual será el próximo candidato a escoger (ya sea para agregar al modelo en FSS o para eliminarlo en BSS), por lo que no se hace necesario el "iterar en cada iteración" para encontrar el mejor/peor candidato cada vez, ahorrando valioso tiempo de cómputo.

Finalmente, mencionar que ámbas funciones retornan listas para los Errores Cuadráticos Medios tanto de training como de test.

```python
def fss_con_MSI(x_train, y_train, x_test, y_test, candidates):
    mse_train=[]
    mse_test=[]
    c=0
    actual_cand=["intercept"]
    while c<9:
        model = lm.LinearRegression(fit_intercept=False)
        aux_xt=x_train.loc[:,actual_cand].as_matrix()
        aux_xtest=x_test.loc[:,actual_cand].as_matrix()
        mod=model.fit(aux_xt, y_train)
        
        predictions_train = mod.predict(aux_xt)
        predictions_test = mod.predict(aux_xtest)
        
        residuals_train = predictions_train - y_train
        residuals_test = predictions_test - y_test
        mse_dataerror = np.mean(np.power(residuals_train, 2))
        mse_dataerror_test = np.mean(np.power(residuals_test, 2))
        
        mse_train.append(mse_dataerror)
        mse_test.append(mse_dataerror_test)
        actual_cand.append(candidates[c])
        c+=1
    return mse_train,mse_test
```

```python
def bss_con_MSI(x_train, y_train, x_test, y_test, candidates):
    mse_train=[]
    mse_test=[]
    while len(candidates)>1:
        model = lm.LinearRegression(fit_intercept=False)
        aux_xt=x_train.loc[:,candidates].as_matrix()
        aux_xtest=x_test.loc[:,candidates].as_matrix()
        mod=model.fit(aux_xt, y_train)
        
        predictions_train = mod.predict(aux_xt)
        predictions_test = mod.predict(aux_xtest)
        
        residuals_train = predictions_train - y_train
        residuals_test = predictions_test - y_test
        mse_dataerror = np.mean(np.power(residuals_train, 2))
        mse_dataerror_test = np.mean(np.power(residuals_test, 2))
        
        mse_train.append(mse_dataerror)
        mse_test.append(mse_dataerror_test)
        candidates.pop(-2)
    return mse_train,mse_test
```

Se convierten los Dataframes de entrenamiento y prueba a Matrices para un mejor manejo, y con estos datos se invoca a las funciones FSS y BSS, obteniendo así listas con los Errores cuadráticos medios de test y training para cada uno.

Además, se extrae la lista de candidatos ordenados por su Mutual Information Score de la función MIS_order(), lista que servirá como parámetro para las funciones fss_con_MSI/bss_con_MSI, las cuales tambien retornan listas con los errores respectivos

```python
names_regressors = ["Lcavol", "Lweight", "Age", "Lbph", "Svi", "Lcp", "Gleason", "Pgg45"]
Xm = Xtrain.as_matrix()
ym = ytrain.as_matrix()
Xt = Xtest.as_matrix()
yt = ytest.as_matrix()

candidate_order = MIS_order()
cand_list = [ str(i[0]) for i in candidate_order ]
cand_list2=cand_list[:]

a,mse_train_fss1, mse_test_fss1=fss(Xm,ym,Xt,yt,names_regressors)
b,mse_train_bss1, mse_test_bss1=bss(Xm,ym,Xt,yt,names_regressors)

mse_train_fss2,mse_test_fss2=fss_con_MSI(Xtrain,ytrain.as_matrix(),Xtest,ytest.as_matrix(),cand_list)
mse_train_bss2,mse_test_bss2=bss_con_MSI(Xtrain,ytrain.as_matrix(),Xtest,ytest.as_matrix(),cand_list2)
```

Finalmente, se grafican los cuatro casos ya mencionados.

```python
plot_errors(mse_train_fss1,mse_test_fss1,"Errores FSS normal.")
plot_errors(mse_train_fss2,mse_test_fss2,"Errores FSS con ordenamiento por MIS.")
plot_errors(mse_train_bss1, mse_test_bss1, "Errores BSS normal")
plot_errors(mse_train_bss2, mse_test_bss2, "Errores BSS con ordenamiento por MIS")
```

De los cuatro gráficos anteriores, se puede concluir que el criterio de ordenamiento por Mutual Information Score es de mucha utilidad, ya que aunque posea mayores errores en general, la diferencia de errores no es tan significativa con respecto al beneficio de usar el criterio, el cual permite "ahorrar" iteraciones cada vez que se quiera encontrar al mejor candidato, ya que al estar ya ordenados estos candidatos de antemano, no es necesario encontrarlos.

Se puede observar ademas que el número óptimo de variables ajustadas en el modelo, el cual permite minimizar el error cuadrático medio para los datos de prueba, se encuentra entre las 3 y 4 variables. Este caso se repite en los cuatro gráficos anteriores.
